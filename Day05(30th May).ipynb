### Draw a pie plot based on number of wins for each team


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plotter
class wins():
    def winlist(self):
        df=pd.read_csv('DataFiles/Match.csv')
        teams=[]
        t=[]
        for row in df.values:
            t.append(row[10])
        teams,counts=np.unique(t,return_counts=True)
        list={'Teams':teams,'wincount':counts}
        df1=pd.DataFrame(list)
        pieLabels = ['Chennai Super kings', 'Deccan Chargers', 'Delhi DareDevils', 'Gujarat lions', 'Kings XI Punjab', 
        'Kochi Tuskers Kerala','Kolkata Knight Riders','Mumbai Indians','Pune Warriors','Rajasthan Royals',
        'Rising Pune Giants','Royal Challengers Banglore','Sun Risers Hyderabad','abandoned','Tied','nan']
        populationShare     = counts
        plotter.pie(populationShare,labels=pieLabels,autopct='%1.2f',startangle=90)
        plotter.show()
mc=wins()
mc.winlist()
```

### Draw a bar graph for number of matches won (i,e.consider win if the team wins both the toss and match)  for the team “Mumbai Indians” for all years


```python
import pandas as pd
import numpy as np
class my_class():
    def mumbaiindians(self):
        df=pd.read_csv('DataFiles/Match.csv')
        t=[]
        teams=[]
        winlist=0
        for row in df.values:
            t.append(row[9])
        years=[]
        for row in df.values:
            years.append(row[5])
        year=np.unique(years)
        win=[]
        for i in year:
            winlist=0
            for row in df.values:
                if(i==row[5])and (row[9]=='Mumbai Indians') and (row[10]=='Mumbai Indians'):
                    winlist=winlist+1
            win.append(winlist)
        x=year
        y=win
        width=1/2
        plotter.bar(x,y,width,color='green')
        plotter.show()
mi=my_class()
mi.mumbaiindians()
```
