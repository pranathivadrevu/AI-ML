```python
import pandas as pd
import numpy as np
def matchwonfunction(filename):
    df=pd.read_csv(filename)
    a=df.groupby(['match_winner','Season_Year']).match_winner.count()
    a=a.unstack(level=1).fillna('0')
    return a
matchwonfunction('DataFiles/Match.csv')
```

```python
import pandas as pd
import numpy as np
def manofmatch(filename):
    df=pd.read_csv(filename)
    mem=[]
    manofmatch=[]
    for row in df.values:
        if(row[7]=='Hyderabad'):
            mem.append(row[14])
    manofmatch,counts=np.unique(mem,return_counts=True) 
    list={"Player":manofmatch,"ManOfMatchCounts":counts}
    df1=pd.DataFrame(list)
    return df1
manofmatch('DataFiles/Match.csv')
```

```python
 Print number of matches won and loss(as two columns). Consider as win when a team wins Toss  and Match . print for each and every team
```

```python
import pandas as pd
import numpy as np
def winloss(filename):
    df=pd.read_csv(filename)
    teams=[]
    t=[]
    for row in df.values:
        t.append(row[9])
    teams=np.unique(t)    
    winlist=[]
    losslist=[]
    for i in teams:
        wc=0
        lc=0
        for row in df.values:
            if(i==row[9]) and (i==row[10]) :
                wc=wc+1
            elif(row[2]==i or row[3]==i):
                lc=lc+1
        losslist.append(lc)
        winlist.append(wc)
    data={'teams':teams,"win":winlist,'loss':losslist}      
    df1=pd.DataFrame(data)
    return df1
winloss('DataFiles/Match.csv')

```
